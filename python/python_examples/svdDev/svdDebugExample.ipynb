```python
import pyJvsip as pv
import vsipUser as vu
import svdUtils as u
import decompositionUtilities as du
A=pv.create('cmview_f',7,5)
A.realview.block.vector.randn(5)
A.imagview.block.vector.randn(7)
A.mprint('%.2f')
```

```python
def chouseProd_f(v, A):
    vu.mprint(v,'%.4f')
    a_atr=pv.vsip_cmattr_f()
    beta = 2.0/pv.vsip_cvjdot_f(v,v).r
    print('beta %.4f'%beta)
    pv.vsip_cmgetattrib_f(A,a_atr)
    B=pv.vsip_cmcreate_f(a_atr.col_length,a_atr.row_length,pv.VSIP_ROW,pv.VSIP_MEM_NONE)
    w = pv.vsip_cvcreate_f(a_atr.row_length,pv.VSIP_MEM_NONE)
    pv.vsip_cvconj_f(v,v)
    pv.vsip_cvmprod_f(v,A,w);pv.vsip_cvconj_f(w,w)
    pv.vsip_cvconj_f(v,v)
    print('w');vu.mprint(w,'%.4f')
    pv.vsip_cvouter_f(pv.vsip_cmplx_f(beta,0.0),v,w,B)
    pv.vsip_cmsub_f(A,B,A)
    pv.vsip_cvalldestroy_f(w)
    pv.vsip_cmalldestroy_f(B)
```

```python
def houseProd(v,A):
    v.mprint('%.4f')
    beta = 2.0/v.jdot(v)
    print('beta (%.4f,%.4f)'%(beta.real,beta.imag))
    v.conj;w=v.prod(A).conj;v.conj
    print('w');w.mprint('%.4f')
    A -= v.outer(beta,w)
    return A
```

```python
def cbidiag_f(a):
    m = pv.vsip_cmgetcollength_f(a)
    n = pv.vsip_cmgetrowlength_f(a)
    B = u.cmclone_f(a)
    Bs= pv.vsip_cmcloneview_f(B)
    x=pv.vsip_cmcolview_f(B,0)
    v=u.cvclone_f(x)
    vs = pv.vsip_cvcloneview_f(v)
    for i in range(n-1):
        pv.vsip_cvputlength_f(v,m-i)
        pv.vsip_cvcopy_f_f(u.ccol_sv_f(u.cmsv_f(B,Bs,i,i),x,0),v)
        u.chouseVector_f(v)
        pv.vsip_csvmul_f(pv.vsip_cdiv_f(pv.vsip_cmplx_f(1.0,0.0),pv.vsip_cvget_f(v,0)),v,v)
        chouseProd_f(v,Bs)
        vu.mprint(B,'%.4f')
        pv.vsip_cvcopy_f_f(u.cvsv_f(v,vs,1),u.cvsv_f(x,x,1))
        if i < n-2:
            j = i+1
            pv.vsip_cvputlength_f(v,n-j)
            pv.vsip_cvcopy_f_f(u.crow_sv_f(u.cmsv_f(B,Bs,i,j),x,0),v)
            u.chouseVector_f(v);pv.vsip_cvconj_f(v,v)
            pv.vsip_csvmul_f(pv.vsip_cdiv_f(pv.vsip_cmplx_f(1.0,0.0),pv.vsip_cvget_f(v,0)),v,v)
            u.cprodHouse_f(Bs,v)
            pv.vsip_cvcopy_f_f(u.cvsv_f(v,vs,1),u.cvsv_f(x,x,1))
    if m > n:
        i=n-1
        pv.vsip_cvputlength_f(v,m-i)
        pv.vsip_cvcopy_f_f(u.ccol_sv_f(u.cmsv_f(B,Bs,i,i),x,0),v)
        u.chouseVector_f(v)
        pv.vsip_csvmul_f(pv.vsip_cdiv_f(pv.vsip_cmplx_f(1.0,0.0),pv.vsip_cvget_f(v,0)),v,v)
        chouseProd_f(v,Bs)
        pv.vsip_cvcopy_f_f(u.cvsv_f(v,vs,1),u.cvsv_f(x,x,1))
    pv.vsip_cvdestroy_f(vs); pv.vsip_cvalldestroy_f(v)
    pv.vsip_cvdestroy_f(x); pv.vsip_cmdestroy_f(Bs)
    return B
```

```python
def bidiag(A): # m >= n
    B=A.copy
    m=B.collength
    n=B.rowlength
    for i in range(n-1):
        x=B[i:,i:].colview(0)
        v=du.houseVector(x)
        v /= v[0]
        houseProd(v,B[i:,i:])
        B.mprint('%.4f')
        x[1:]=v[1:]
        if i < n-2:
            j=i+1
            x = B[i:,j:].rowview(0)
            v=du.houseVector(x).conj 
            v /= v[0]
            du.prodHouse(B[i:,j:],v)#=B[i:,j:].prod(house(v))
            x[1:]=v[1:]
    if m > n: #do last column if matrix not square
        i=n-1
        x=B[i:,i:].colview(0)
        v=du.houseVector(x)
        v /= v[0]
        houseProd(v,B[i:,i:])
        x[1:]=v[1:]
    return B
```
