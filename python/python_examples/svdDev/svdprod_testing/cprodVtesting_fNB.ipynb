```python
print('case LSIDE, m < n; FULL decomposition')
m=4;n=6
A=pv.create('cmview_f',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svd
B=v.empty.fill(0.0)
B.diagview(0).fill(1)
B.mprint(f)
d=pv.SV('csv_f',m,n,'FULL','FULL')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
v.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-v).normFro)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B.view)
print('should be identity matrix')
B.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
v.herm.mprint(f)
print('Frobenius Norm %.4e'%(B-v.herm).normFro)
```

```python
print('case RSIDE, m > n; FULL decomposition')
m=6;n=4
A=pv.create('cmview_f',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svd
B=pv.create(v.type,v.collength,v.collength).fill(0.0)
B.diagview(0).fill(1)
B.mprint(f)
d=pv.SV('csv_f',m,n,'FULL','FULL')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_RSIDE,B.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
v.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-v).normFro)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('should be identity matrix')
B.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
v.herm.mprint(f)
print('Frobenius Norm %.4e'%(B-v.herm).normFro)
```

```python
print('case LSIDE, m < n; PART decomposition')
m=4;n=6
A=pv.create('cmview_f',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svdP
B0=v.empty.fill(0.0)
B=B0[:v.rowlength,:]
B.diagview(0).fill(1.0)
B.mprint(f)
d=pv.SV('csv_f',m,n,'PART','PART')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be the same; norm of difference should be small')
B0.mprint(f)
v.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B0-v).normFro)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B0.view)
print('should be identity matrix')
B.mprint(f)
B1=pv.create(v.type,v.collength,v.collength).fill(0.0)
B1.diagview(0).fill(1.0)
B=B1[:v.rowlength,:]
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B1.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
v.herm.mprint(f)
print('Frobenius Norm %.4e'%(B-v.herm).normFro)
```

```python
print('case RSIDE, m < n; PART decomposition')
m=4;n=6
A=pv.create('cmview_f',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svdP
B0=pv.create(v.type,v.collength,v.collength).fill(0.0)
B0.diagview(0).fill(1)
B=B0[:,:v.rowlength]
B0.mprint(f)
d=pv.SV('csv_f',m,n,'PART','PART')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_RSIDE,B0.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
v.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-v).normFro)
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('should be the same')
B0.mprint(f)
(v.prod(v.herm)).mprint(f)
B1=B0[:v.rowlength,:].fill(0.0);B1.diagview(0).fill(1.0)
B=B1[:,:v.rowlength]
pv.vsip_csvdprodv_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('\nshould be the same; norm of difference should be small')
B1.mprint(f)
v.herm.mprint(f)
print('Frobenius Norm %.4e'%(B1-v.herm).normFro)
```
