>Cases SVD FULL   
>>+ $M < N$; NTRANS; RSIDE   
>>+ $M < N$; NTRANS; LSIDE   
>>+ $M < N$; HERM; RSIDE   
>>+ $M < N$; HERM; LSIDE   
>>+ $M \geq N$; NTRANS; RSIDE   
>>+ $M \geq N$; NTRANS; LSIDE   
>>+ $M \geq N$; HERM; RSIDE   
>>+ $M \geq N$; HERM; LSIDE   

>Cases SVD PART   
>>+ $M < N$; NTRANS; RSIDE   
>>+ $M < N$; NTRANS; LSIDE   
>>+ $M < N$; HERM; RSIDE   
>>+ $M < N$; HERM; LSIDE   
>>+ $M \geq N$; NTRANS; RSIDE   
>>+ $M \geq N$; NTRANS; LSIDE   
>>+ $M \geq N$; HERM; RSIDE   
>>+ $M \geq N$; HERM; LSIDE  

```python
print('case RSIDE, m < n; FULL decomposition')
m=4;n=6
A=pv.create('cmview_f',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svd
B=u.empty.fill(0.0)
B.diagview(0).fill(1)
B.mprint(f)
d=pv.SV('csv_f',m,n,'FULL','FULL')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodu_f(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_RSIDE,B.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
u.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-u).normFro)
pv.vsip_csvdprodu_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('should be identity matrix')
B.mprint(f)
pv.vsip_csvdprodu_f(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
u.herm.mprint(f)
print('Frobenius Norm %.4e'%(B-u.herm).normFro)
```

For cases below matrix U will no longer be square. This means the output matrix is no longer the same size as the input matrix which complicates the problem of designing a data space with a suitable data footprint suitable for both input and output.

```python
print('case RSIDE, m > n; PART decomposition')
m=6;n=4
A=pv.create('cmview_d',m,n).randn(5)
f='%.3f'
A.mprint(f)
u,s,v=A.copy.svdP
B0=pv.create(u.type,u.collength,u.collength).fill(0.0)
B=B0[:,:u.rowlength]
B0.diagview(0).fill(1)
B0.mprint(f)
d=pv.SV('csv_d',m,n,'PART','PART')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodu_d(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_RSIDE,B0.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
u.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-u).normFro)
pv.vsip_csvdprodu_d(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_RSIDE,B.view)
print('shouldbe the same; norm of differences should be small')
print('result')
B0.mprint(f)
print('u.prod(u.herm) direct');u.prod(u.herm).mprint(f)
print('Frobenius Norm %.4e\n\n'%(B0-u.prod(u.herm)).normFro)
```

```python
print('case LSIDE, m > n; PART decomposition')
m=6;n=4
A=pv.create('cmview_d',m,n).randn(5)
f='%.3f'
print('input matrix and initial output matrix')
A.mprint(f)
u,s,v=A.copy.svdP
B0=pv.create(u.type,u.collength,u.rowlength).fill(0.0)
B=B0[:u.rowlength,:]
B.diagview(0).fill(1)
B0.mprint(f)
d=pv.SV('csv_d',m,n,'PART','PART')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_csvdprodu_d(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be the same; norm of difference should be small')
B0.mprint(f)
u.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B0-u).normFro)
pv.vsip_csvdprodu_d(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B0.view)
print('should be identity matrix')
B.mprint(f)
B1=pv.create(u.type,u.collength,u.collength).fill(0.0)
B1.diagview(0).fill(1.0)
B=B1[:u.rowlength,:]
pv.vsip_csvdprodu_d(d.vsip,pv.VSIP_MAT_HERM,pv.VSIP_MAT_LSIDE,B1.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
u.herm.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-u.herm).normFro)
```
