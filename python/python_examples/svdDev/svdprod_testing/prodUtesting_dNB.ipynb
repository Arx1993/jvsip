```python
import pyJvsip as pv
```

```python
print('case LSIDE, m < n; FULL decomposition')
m=4;n=6
A=pv.create('mview_d',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svd
B=u.empty.fill(0.0)
B.diagview(0).fill(1)
B.mprint(f)
d=pv.SV('sv_d',m,n,'FULL','FULL')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be the same; norm of difference should be small')
B.mprint(f)
u.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B-u).normFro)
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_TRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be identity matrix')
B.mprint(f)
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_TRANS,pv.VSIP_MAT_LSIDE,B.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
u.transview.mprint(f)
print('Frobenius Norm %.4e'%(B-u.transview).normFro)
```

```python
# U B; U^T B; U^T B; U is 6x4 B is 4x4 out is 6x4; U^T is 4x6, B is 6x6 out is 4 by 6;
print('case LSIDE, m > n; PART decomposition')
m=6;n=4
A=pv.create('mview_d',m,n).randn(5)
f='%.3f'
print('Test matrix')
A.mprint(f)
u,s,v=A.copy.svdP
B0=pv.create(u.type,6,4).fill(0.0)
B=B0[:u.rowlength,:]
B.diagview(0).fill(1)
B.mprint(f)
d=pv.SV('sv_d',m,n,'PART','PART')
S=s.empty.fill(0.0)
_=d.svd(A.copy,S)
print('should be the same')
S.mprint(f)
s.mprint(f)
#U is 6x4 B is 4x4 out is 6x4;
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_NTRANS,pv.VSIP_MAT_LSIDE,B.view)
print('should be the same; norm of difference should be small')
B0.mprint(f)
u.mprint(f)
print('Frobenius Norm %.4e\n\n'%(B0-u).normFro)
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_TRANS,pv.VSIP_MAT_LSIDE,B0.view)
print('should be identity matrix')
B.mprint(f)
B0=pv.create(u.type,u.collength,u.collength).fill(0.0)
B0.diagview(0).fill(1.0)
B=B0[:u.rowlength,:]
pv.vsip_svdprodu_d(d.vsip,pv.VSIP_MAT_TRANS,pv.VSIP_MAT_LSIDE,B0.view)
print('\nshould be the same; norm of difference should be small')
B.mprint(f)
u.transview.mprint(f)
print('Frobenius Norm %.4e'%(B-u.transview).normFro)
```
