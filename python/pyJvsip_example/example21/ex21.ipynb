```python
import pyJvsip as pv
from math import pi as M_PI
from math import cos as cos
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scale gram to db, min 0 max 255 */
def scale(gram_data,gram):
    pv.cmagsq(gram_data,gram)
    gram += (1.0-gram.minval)
    gram.log10
    gram *= 256.0/gram.maxval
    return gram
```

```python
def center(gram):
    assert gram.rowstride == 1
    assert gram.colstride == gram.rowlength
    if gram.collength & 1: #odd
        for i in range(gram.rowlength):
            pv.freqswap(gram.colview(i))
    else: #even number of columns; use trick
        lngth = int(gram.collength*gram.rowlength/2)
        gramvec=gram.block.vector
        pv.swap(gramvec[lngth:],gramvec[:lngth])     
    return gram
```

```python
def view_store(M,fname): #VU_mprintgram
    assert 'pyJvsip.__View' in repr(M),'First input argument must be a pyJvsip view'
    assert M.type in ['vview_f','mview_f','vview_d','mview_d'], 'View type not supported'
    assert isinstance(fname,str), 'File name is a string'
    if 'mview' in M.type:
        RL=M.rowlength; CL=M.collength
        of=open(fname,'w')
        of.write('%s\n'%M.type)
        of.write('%ld %ld\n'%(CL,RL))
        for row in range(CL):
            for col in range(RL):
                of.write('%ld %ld %e\n'%(row,col,M[row,col]))
        of.close()
    else:
        L=M.length
        of=open(fname,'w')
        of.write('%s'%M.type)
        of.write('%ld'%L)
        for i in range(L):
            of.write('%ld %e'%(i,M[i]))
        of.close()
    return
```

```python
def view_read(fname):
    from pyJvsip import create as create
    fd=open(fname,'r')
    t=fd.readline().split()[0]
    assert t in ['vview_f','mview_f','vview_d','mview_d'], 'Type <:%s:> not supported'%t
    if 'mview' in t:
        sz=fd.readline().split()
        M=pv.create(t,int(sz[0]),int(sz[1]))
        for lin in fd:
            a=lin.split()
            r=int(a[0]);c=int(a[1]);x=float(a[2])
            M[r,c]=x
    else:
        sz=fd.readline().split()
        M=create(t,int(sz))
        for lin in fd:
            a=lin.split()
            i=int(a[0]);x=float(a[2])
            M[i]=x
    fd.close()
    return M
```

```python
# program parameters
D=1.5        # Sensor Spacing Meters
Fs=1000.0    # Sample Frequency (Hz)
F0=450.0     # Target Frequency
F1=300.0     # Target Frequency
F2=150.0     # Target Frequency
F3=50.0      # Target Frequency
Theta_o=40 # Target Direction
Ns=512       # length of sampled time series (samples)
Nn=1024      # length of (simulated) noise series
Mp=128       # number of sensors in linear array
c=1500.0     # Propagation Speed (Meters/Second)
Nnoise=64    # number of Simulated Noise Directions
kaiser=9     # window parameter
Nfilter=10   # Kernel length for noise filter
```

```python
windowt=pv.create('vview_d',Ns).hanning
windowp=pv.create('vview_d',Mp).hanning
kernel=pv.create('vview_d',Nfilter).kaiser(kaiser)
fir= pv.FIR('fir_d',kernel,'NONE',2 * Nn,2,'YES');
t =pv.create('vview_d',Ns).ramp(0,1)
st=pv.Rand('PRNG',15)
#vsip_randstate *state = vsip_randcreate(15,1,1,VSIP_PRNG);
noise=pv.create('mview_d',Nnoise,Nn).fill(0.0)
nv = pv.create('vview_d',2 * Nn)
tt = pv.create('vview_d',Ns)
data =pv.create('mview_d',Mp,Ns,'ROW')
gram_data = pv.create('cmview_d',Mp,Ns/2 + 1,'COL')
gram = pv.create('mview_d',Mp,Ns/2 + 1,'ROW')
Xim = pv.create('mview_d',Mp,Mp+1,'ROW')
alpha = (D * Fs) / c
m = pv.create('vview_d',Mp).ramp(0,1)
Xi = pv.create('vview_d',Mp + 1).ramp(0,M_PI/Mp).cos
w0 = 2 * M_PI * F0/Fs
w1 = 2 * M_PI * F1/Fs
w2 = 2 * M_PI * F2/Fs
w3 = 2 * M_PI * F3/Fs
cnst1 = M_PI/Nnoise
offset0 = int(alpha * Mp + 1)
rcfftmop = pv.FFT('rcfftmop_d',(Mp,Ns,1,pv.VSIP_ROW,0,0))
ccfftmip = pv.FFT('ccfftmip_d',(Mp,Ns/2 + 1,pv.VSIP_FFT_FWD,1,pv.VSIP_COL,0,0))
pv.outer(alpha,m,Xi,Xim)
gram.fill(0.0)
```

```python
Xim[0:4,0:4].mprint('%.3f')
# Simulate some data
for j in range(Nnoise):
    st.randn(nv)
    nr=noise.rowview(j)
    fir.flt(nv,nr)
    nr *= 12.0/(Nnoise) #scale noise
noise.putrowlength(Ns)
for i in range(Mp):
    Xim_val = Xim[i,Theta_o]
    data_v = data.rowview(i)
    pv.ma(t,w0,-w0 * Xim_val,tt)
    pv.cos(tt,data_v)
    pv.ma(t,w1,-w1 * Xim_val,tt)
    data_v += tt.cos
    pv.ma(t,w2,-w2 * Xim_val,tt) 
    data_v += tt.cos #F2 time series */
    pv.ma(t,w3,-w3 * Xim_val,tt)
    tt.cos #F3 time series */
    tt *= 3.0 # scale by 3.0 */
    data_v += tt
    data_v *= 3.0 # scale by 3.0 */
    for j in range(Nnoise):
        nv=noise.rowview(j)
        nv.putoffset(offset0 + int( i * alpha * cos(j * cnst1)))
        data_v += nv
# window the data and the array to reduce sidelobes
pv.mmul(windowt.ROW,data,data)
pv.mmul(windowp.COL,data,data)    
# do ffts 
rcfftmop.dft(data,gram_data)
ccfftmip.dft(gram_data)
gram = scale(gram_data,gram)
center(gram)
view_store(gram,'gram_pyoutput')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.imshow(gram.list,origin='lower')
ax.set_yticks([0,31,63,95,127])
ax.set_xticks([0,51,103,154,206,256])
ax.set_yticklabels([r'$-\hat{k}$','-k/2','0','k/2',' k '])
ax.set_xticklabels([' 0 ','100','200','300','400','500'])
ax.set_title(r'$\vec{k}\omega$ plot')
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Wave Number')
```
