```python
import pyJvsip as pjv
from math import pi as pi
n=1000
x=pjv.create('vview_d',n).ramp(0,2*pi/float(n))
c=pjv.cos(x,x.empty)
cClip=pjv.clip(c,-.9,.9,-.8,.8,c.empty)
```

```python
%matplotlib inline
from matplotlib.pyplot import *
```

```python
subplot(2,1,1)
plot(x,c);title('Sine')
subplot(2,1,2)
plot(x,cClip);title('Cliped Sine')
tight_layout()
savefig('eXclip.pdf',figsize=(3,5))
```

```python
cinvClip=pjv.invclip(c,-.6,0.0,.6,-.7,.7,c.empty)
subplot(2,1,1)
plot(x,c);title('Sine')
subplot(2,1,2)
plot(x,cinvClip);title('Inverse Clip Sine')
tight_layout()
savefig('eXinvclip.pdf',figsize=(3,5))
```
