```python
import pyJvsip as pjv
%matplotlib inline
import matplotlib.pyplot as plt
```

## Window Creation
In C VSIPL windows have create funtions.  This caused some  problems for pyJvsip as I did not want to do a seperate class for every window.  Consequently the baseline window creation is defined using the Block class as shown in the examples below.  Also demonstrated is how to use the create function to produce windows.

At the bottom I show **view** methods for creating windows.

```python
plt.hold=True
plt.plot(pjv.create('kaiser_f',10,9).list,'r')
plt.plot(pjv.Block('kaiser_d',10,9).w.list,'v')
plt.hold=False
```

```python
plt.hold=True
plt.plot(pjv.create('cheby_f',30,50).list,'r')
plt.plot(pjv.Block('cheby_d',30,50).w.list,'v')
plt.hold=False
```

```python
plt.hold=True
plt.plot(pjv.create('blackman_f',30).list,'r')
plt.plot(pjv.Block('blackman_d',30).w.list,'v')
plt.hold=False
```

```python
plt.hold=True
plt.plot(pjv.create('hanning_f',30).list,'r')
plt.plot(pjv.Block('hanning_d',30).w.list,'v')
plt.hold=False
```

```python
a = pjv.create('hanning_f',150)
y=a.fftop
pjv.freqswap(y)
z=y.copy.cmagsq
plt.plot(z[50:100].list)
```

#### View window methods
For vector views one can create a window of the proper length and copy it into the vector using a method as shown below. For **blackman** and **hanning** windows the method is defined as a property as no addtional information is required. For **kaiser** the tansition width parameter is supplied as an argument, and for **cheby** the sidelobe dB *ripple* parameter is supplied as an argument.

```python
a=pjv.create('vview_d',150)
plt.plot(a.blackman.list,'c')
plt.hold = True
plt.plot(a.hanning.list,'r')
plt.plot(a.kaiser(9).list,'b')
plt.plot(a.cheby(50).list,'y')
plt.hold = False
```
